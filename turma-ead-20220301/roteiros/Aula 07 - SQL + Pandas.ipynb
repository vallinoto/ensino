```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///dados/getninjas.db')
```

```python
conn = engine.connect()
```

```python
import pandas as pd
```

```python
dados_getninjas = pd.read_sql_table('users', conn)
```

```python
dados_getninjas.head()
```

```python
dados_req = pd.read_sql_table('requests', conn)
```

```python
dados_req.head()
```

```python
dados_cat = pd.read_sql_table('categories', conn)
```

```python
dados_cat.head()
```

```python
dados_all = pd.merge(dados_req, dados_cat, left_on='category_id', right_on='id')
```

```python
dados_all.head()
```

```python
consulta = """
            SELECT c.name as nome_categoria, count(*) as qtd_pedidos
            FROM requests r
            INNER JOIN categories c
            ON r.category_id = c.id
            GROUP BY c.name
            ORDER BY qtd_pedidos DESC
"""
```

```python
pd.read_sql_query(consulta, conn)
```
