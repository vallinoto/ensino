Barras (dados categóricos)

```python
lista_produto = ['Martelo', 'Prego', 'Parafuso']
val_produto = [10,25,60]
```

```python
plt.bar(lista_produto, val_produto)
plt.show()
```

```python
qtd_pedidos_prodA = [10,32,7,25]
qtd_pedidos_prodB = [3,20,17,51]
```

```python
x1 = np.arange(len(qtd_pedidos_prodA))
x2 = [ x + 0.25 for x in x1 ]
```

```python
plt.bar(x1, qtd_pedidos_prodA, width=0.25, color = 'r')
plt.bar(x2, qtd_pedidos_prodB, width=0.25, color = 'b')

trimestres = ['Trim1', 'Trim2', 'Trim3', 'Trim4']

plt.xticks([ x + 0.25 for x in range(len(qtd_pedidos_prodA)) ], trimestres)

plt.show()
```

```python
plt.bar(x1, qtd_pedidos_prodA, width=0.25, color = 'r')
plt.bar(x2, qtd_pedidos_prodB, width=0.25, color = 'b')

trimestres = ['Trim1', 'Trim2', 'Trim3', 'Trim4']

plt.xticks([ x + 0.25 for x in range(len(qtd_pedidos_prodA)) ], trimestres)

plt.legend(['Prod A', 'Prod B'])

plt.show()
```

```python
plt.bar(x1, qtd_pedidos_prodA, width=0.25, color = 'r')
plt.bar(x2, qtd_pedidos_prodB, width=0.25, color = 'b')

trimestres = ['Trim1', 'Trim2', 'Trim3', 'Trim4']

plt.xticks([ x + 0.25 for x in range(len(qtd_pedidos_prodA)) ], trimestres)

plt.legend(['Prod A', 'Prod B'], loc='upper center')
plt.title('Comparativo entre produto A e B ao longo dos trimestres')

plt.show()
```

```python
plt.style.use('ggplot')

x = [4,6,8,10]
y = [10,20,30,40]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,8))

ax1.plot(x, y, 'o')
ax2.plot(x, y, color='red')

plt.show()
```

```python
plt.style.use('grayscale')

x = [4,6,8,10]
y = [10,20,30,40]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,8))

ax1.plot(x, y, 'o')
ax2.plot(x, y, color='red')

plt.show()
```
