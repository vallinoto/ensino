```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=df_tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=df_tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', data=df_tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', data=df_tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='sex', size='size', sizes=(15,300), data=df_tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='day')
```
