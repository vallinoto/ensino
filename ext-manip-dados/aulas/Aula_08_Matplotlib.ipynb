```
lista_produtos = [  'martelo', 'prego', 'parafuso']
valor_produtos = [15,2,4]
```

```
plt.bar(lista_produtos, valor_produtos)
plt.show()
```

```
plt.bar(lista_produtos, valor_produtos, color='#345b70')
plt.title('Quantidade de Produtos')
plt.show()
```

```
import numpy as np

qtdA = [ 10, 36, 24, 8 ]
qtdB = [14,25,16,15]

trimestre = ['Trim 1','Trim 2','Trim 3','Trim 4']
```

```
plt.bar(np.arange(len(qtdA)), qtdA)
plt.bar([ x + 0.25 for x in range(len(qtdA))], qtdB)
plt.xticks([ x + 0.25 for x in range(len(qtdA))], trimestre)
```

```
plt.bar(np.arange(len(qtdA)), qtdA, width=0.25)
plt.bar([ x + 0.25 for x in range(len(qtdA))], qtdB, width=0.25)
plt.xticks([ x + 0.25 for x in range(len(qtdA))], trimestre)
plt.show()
```

```
trimestre =  ['Trimestre 1','Trimestre 2','Trimestre 3','Trimestre 4']

plt.bar(np.arange(len(qtdA)), qtdA, width=0.25)
plt.bar([ x + 0.25 for x in range(len(qtdA))], qtdB, width=0.25)
plt.xticks([ x + 0.25 for x in range(len(qtdA))], trimestre)
plt.xticks(rotation=45)
plt.legend(['Produto A','Produto B'])
plt.show()
```
