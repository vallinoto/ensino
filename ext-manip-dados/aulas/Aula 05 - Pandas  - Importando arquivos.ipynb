#### Importando arquivos CSV

Por definição, CSV - comma-separated-values - (valores separados por vírgulas).
Normalmente, arquivos com  essa extensão são separados ou delimitados por uma vírgula.
A grande vantagem de um arquivo CSV é o fato de ele possibilitar a importação e exportação arquivos de uma linguagem que vários aplicativos podem ler.
